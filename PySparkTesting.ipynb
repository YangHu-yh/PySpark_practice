```python
class Parent():
    def __init__(self):
        self._number = 20
```

```python
class Child():
    def __init__(self):
        Parent.__init__(self)

        self._number = 50

    def get_number(self):
        return self._number
    
    def get_parent_number(self):
        Parent.__init__(self)
        return self._number
```

```python
parent1 = Parent()
parent1
```

```python
print(parent1._number)
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, model_name):
        self.model_name = model_name
      
    @abstractmethod
    def move(self):
        pass
          
    def get_model_name(self):
        return self.model_name

class Car(Vehicle):
    def __init__(self, model_name):
          super().__init__(model_name)
          
    
    def move(self):
        return f"Move the {self.model_name}"
    
    def get_model_name(self):
        return self.model_name # Accessing "protected" member


```

```python
my_car = Car("Toyota")
my_car.get_model_name()
```

```python
my_car.move()
```

```python
class Bicycle(Vehicle):
    def __init__(self, model_name):
        super().__init__(model_name)
    
    def move(self):
        print("Moved slowly")
```

```python
my_bike = Bicycle("Test")
```

```python
my_bike.move()
```
